##Prepare data from PharmGKB downloads:

```python
Core_genes = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/ADME Project/Listed Genes.csv")
```

```python
# Read PharmGKB results for all listed extended ADME genes.
file_paths = []

# List of TSV file paths
for i in range(len(Core_genes['Gene'])):
  if Core_genes['Extended?'][i]==0 and Core_genes['Core?'][i]==0:
    gene_name = Core_genes['Gene'][i]
    file_paths.append(f'/content/gdrive/My Drive/Colab Notebooks/ADME Project/{gene_name}.tsv')

# Create an empty list to store DataFrames
dfs = []

# Read each TSV file and append its DataFrame to the list
for file_path in file_paths:
    try:
        df = pd.read_csv(file_path, sep='\t', error_bad_lines=False)
        dfs.append(df)
    except pd.errors.ParserError:
        print(f"Error reading file: {file_path}")

# Concatenate the DataFrames into a single DataFrame
raw_df = pd.concat(dfs, ignore_index=True)

# Save the combined DataFrame as a CSV file
#raw_df.to_csv('/content/gdrive/My Drive/Colab Notebooks/ADME Project/listedCore_auto.csv', index=False)
```

```python
# Ref/Alter rows creation (Real-run)
data = raw_df

# Define the search patterns and corresponding replacement values
patterns = ['Allele A ', 'Allele C ', 'Allele G ', 'Allele T ']
patterns2 = ['allele A', 'allele C', 'allele G', 'allele T']
replacements = ['A', 'C', 'G', 'T']

# Create the two new columns using the search patterns and replacements
data['Alter'] = data['Association'].apply(lambda x: next((replacements[i] for i, pattern in enumerate(patterns) if pattern in x), np.nan))
data['Ref'] = data['Association'].apply(lambda x: next((replacements[i] for i, pattern2 in enumerate(patterns2) if pattern2 in x), np.nan))

data
```

## Arthitecture and Principal:
This code should be able to rearrange data downloaded from PharmGKB with only two added columns ("Alter" and "Ref"), and produces a usable datasheet with chormosome, location, alter, ref and a label of pharmacogenetic significance.

```python
# Load data
#data = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/ADME Project/listed_core_o.csv")
```

```python
# Only select SNPs which pharmacogenetic significance is known

data['numeric'] = data['Significance'].map({'yes': 1, 'no': 0})
data_dropped = data.dropna(subset=['numeric','Alter'])#no'Ref'!
```

```python
# Sort with variant, alternative base and significant label
#This sorted data set can refered later for certain variant location
#as it includes all related literature.

data_sorted = data_dropped.sort_values(['Variant','Alter','Significance'])

data_sorted.to_csv('/content/gdrive/My Drive/Colab Notebooks/ADME Project/ListedOthers_o.csv')
```

```python
# If for the same alternative base and same variant location, there are both significant report and non-siginificant report,
#delete the later.

data_unique = data_sorted.drop_duplicates(subset=['Variant', 'Alter'], keep='last')

data_unique
```

```python
import requests

# Create empty lists to store the genomic information
chromosomes = []
locations = []
reference_bases = []

# Base URL for Ensembl REST API
base_url = 'https://rest.ensembl.org'

# Iterate over rsIDs
for rsID in data_unique['Variant']:
    # Construct the API endpoint URL
    endpoint = f'{base_url}/variation/human/{rsID}?phenotypes=1'

    # Send GET request to the API
    response = requests.get(endpoint, headers={'Content-Type': 'application/json'})

    # Check if the request was successful
    if response.status_code == 200:
        # Extract relevant information from the response
        data = response.json()
        chromosome = data['mappings'][0]['seq_region_name']
        location = data['mappings'][0]['start']
        reference_base = data['mappings'][0]['allele_string'].split('/')[0]

        # Append the genomic information to the respective lists
        chromosomes.append(chromosome)
        locations.append(location)
        reference_bases.append(reference_base)
    else:
        # If there was an error, append None to the lists
        chromosomes.append(None)
        locations.append(None)
        reference_bases.append(None)

# Create a new DataFrame with the genomic information
search_df = pd.DataFrame({
    'Chromosome': chromosomes,
    'Location': locations,
    'Reference Base': reference_bases
})
```

```python
# Combine the search results with orginal dataset
search_df.set_index(data_unique.index, inplace=True)
combined_df = data_unique.join(search_df)
combined_df
```

```python
# Filter duplicates
combined_df['Combination'] = combined_df[['Alter', 'Ref']].apply(lambda x: ''.join(sorted(x.values.astype(str))), axis=1)
# Drop rows where Column1 and Combination values are the same, and Column4 is 'no'
filterd_df = combined_df.dropna(subset=['Chromosome','Location','Reference Base'])
filterd_df = filterd_df[~((combined_df.duplicated(subset=['Location', 'Combination']) & (combined_df['numeric'] == 0)) |
      ((combined_df['numeric'] == 1) & combined_df.duplicated(subset=['Location', 'Combination']) & (combined_df.groupby('Location')['numeric'].transform('max') == 1)))]
```

```python
# Finalizing
#"Pediatric" deleted due to irrelavance
final1 = filterd_df[['Chromosome','Genes','Location','Reference Base','Alter','Combination','numeric','Literature','Drugs','Association','More Details']]
```

```python
# Filter again
#1)Combination Combination - Reference Base
#2)Delete nan Combination
#3)Change the rest Combination to "True Alter"
filtered2 = final1
filtered2['Combination'] = final1.apply(lambda row: str(row['Combination']).replace(str(row['Reference Base']), ''), axis=1)
```

```python
data_dropped2 = filtered2[~(filtered2['Combination'] == 'nan')]
```

```python
# Final 2: Calculate True Alter
final2 = data_dropped2
final2['Combination'] = data_dropped2.apply(lambda row: row['Alter'] if str(row['Alter']) != str(row['Reference Base']) else str(row['Combination']), axis=1)
```

```python
# Final 3: Delete irrelevant column
final3 = final2.drop(columns='Alter')
final3 = final3.rename(columns={'Combination': 'Altered Base'})
final3 = final3[~(final3['Altered Base'] == None)]
final3
```

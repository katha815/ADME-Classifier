```python
from google.colab import drive
drive.mount('/content/gdrive')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the three datasets
core = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/ADME Project/Core.csv")
extended = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/ADME Project/Extended.csv")
others = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/ADME Project/Others.csv")

# Add a column indicating the source dataset
core['Dataset'] = 'Core'
extended['Dataset'] = 'Extended'
others['Dataset'] = 'Others'

# Merge the datasets into one
comparison = pd.concat([core, extended, others], ignore_index=True)

# Rename the merged dataset to "Comparison"
comparison.name = "Comparison"
comparison
```

```python
# Read the "337SNVs" dataset
snvs = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/ADME Project/337SNVs.csv")

# Add a column to store the check result
comparison['Check Result'] = comparison.apply(lambda row: snvs.loc[
    (snvs['chr'] == row['Chromosome']) & (snvs['Location'] == row['Location']), 'Alter'].values[0]
    if ((snvs['chr'] == row['Chromosome']) & (snvs['Location'] == row['Location'])).any()
    else '', axis=1)
# Check if the alter from old dataset matches with the comparison dataset
comparison['Alter Match'] = comparison.apply(lambda row: snvs.loc[
    (snvs['chr'] == row['Chromosome']) & (snvs['Location'] == row['Location']), 'Alter'].values[0] == row['Check Result']
    if ((snvs['chr'] == row['Chromosome']) & (snvs['Location'] == row['Location'])).any()
    else None, axis=1)
comparison
```

```python
comparison.to_csv('/content/gdrive/My Drive/Colab Notebooks/ADME Project/comparison_results.csv', index=False)
```
